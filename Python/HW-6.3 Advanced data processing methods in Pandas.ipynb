3.1. Створіть нову колонку в даних `Rating Category`, яка визначає категорію рейтингу (`Rating`): високий, середній чи низький за наступним правилом:
- якщо рейтинг вищий або рівний 8 - то він високий (`High`)
- якщо рейтинг від 6 до 8 не включаючи - то він середній (`Medium`)
- інакше - низький (`Low`).

```python
# Method 2

# Step 1. Create a new column 'Rating Category' using conditions
def category_rating(val):
    if val >= 8:
        return 'High'
    elif val >= 6 and val < 8:
        return 'Medium'
    else:
        return 'Low'
        
# Step 2. Apply the function to create the new column
df['Rating Category'] = df.Rating.apply(category_rating)

# Step 3. Display the first ten rows to verify the new column
df[['Rating Category','Rating']].head(10)
```

```python
# Method 2

# Step 1. Create a new column 'Rating Category' using conditions
df['Rating Category'] = pd.cut(df['Rating'],
                              bins=[0, 6, 8, float('inf')],
                              labels=['Low', 'Medium', 'High'],
                              include_lowest=True)

# Step 2. Display the first ten rows to verify the new column
df[['Rating Category','Rating']].head(10)
```

```python
# Method 3
import numpy as np

# Step 1. Define conditions and choices
conditions = [
    df['Rating'] >= 8,
    (df['Rating'] >= 6) & (df['Rating'] < 8),
    df['Rating'] < 6
]
choices = ['High', 'Medium', 'Low']

# Step 2. Create new column using np.select
df['Rating Category'] = np.select(conditions, choices)

# Step 3. Display the first ten rows to verify the new column
df[['Rating Category','Rating']].head(10)
```

**Analysis of Purchase Hours**

After analyzing the number of purchases by hour, we can see that the highest number of purchases occurs at 7 PM (113 = 11,3%) 1 PM (103 = 10,3%), 3 PM (102 = 10.2% ).

This seems logical:

- 1 PM – lunchtime, when many people have a break and can visit a store.
- 3 PM – the active part of the afternoon, possibly related to errands or shopping before the workday ends.
- 7 PM – after working hours, when people are heading home and may stop by a store.

So, the peak purchasing hours are during lunch and after work, which aligns with most people’s typical daily routines. This insight can be used to:

- better schedule staff shifts (increase staff during peak hours),
- plan promotions during off-peak times,
- optimize checkout operations.
